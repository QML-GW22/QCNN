```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2664)
```

```python
x = np.linspace(0, 100, 1000)
noise = np.random.normal(0,1,1000)
x_noise = x + noise
y = np.sin(x/(2*np.pi))
y_noise = np.sin(x_noise/(2*np.pi))
#noise = np.random.normal(0,1,10000)
```

```python
yr=np.random.randn(1000)
plt.plot(x,yr)
```

```python
plt.plot(x,y,alpha=0.5)
plt.plot(x,x_noise,alpha=0.5)
plt.plot(x,y_noise,alpha=0.5)
plt.show()
```
